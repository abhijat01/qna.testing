```python
import os 
import mimetypes 
```

```python
media_dir = os.path.join(os.path.expanduser("~"), 'kb.html', 'media') 
```

```python
onlyfiles = [(f, os.path.join(media_dir,f)) for f in os.listdir(media_dir) if os.path.isfile(os.path.join(media_dir, f))]
```

```python
def mime_guess(file_name, file_path):
    t,e = mimetypes.guess_type(file_path)
    if t is None:
        raise Exception("Could not guess content type for {}".format(file_name))
    print("Content-type:{}, Filename:{}, Encoding:{}".format(t, file_name, e))
```

```python
for f in onlyfiles:
    mime_guess(f[0], f[1])
```
